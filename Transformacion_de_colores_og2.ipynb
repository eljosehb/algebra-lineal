```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Verificación
print("cwd:", os.getcwd())
print("imagen existe?:", os.path.exists("imagen.jpg"))

image = cv2.imread("imagen.jpg")
if image is None:
    raise FileNotFoundError("No se pudo leer imagen.jpg")

original = image.copy()

# Transformación identidad
A = np.eye(3, dtype=np.float32)
c = np.zeros((3,1), dtype=np.float32)

H, W, _ = image.shape

for m in range(H):
    for n in range(W):
        b, g, r = image[m,n]
        v = np.array([[b],[g],[r]], dtype=np.float32)
        u = A @ v + c
        image[m,n] = np.clip(u.T, 0, 255)

# Mostrar con matplotlib
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Transformada")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()

# Guardar
cv2.imwrite("imagen_transformada.jpg", image)
```
