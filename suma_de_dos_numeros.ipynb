# Suma de dos números sin usar +, sum() ni __add__()

Este programa suma dos números solicitados al usuario sin utilizar:
- El operador de suma `+`
- La función `sum()`
- El método `__add__()`

Se presentan varias alternativas para resolver el problema.

```python
## Opción 1: Usando operador - con número negativo

def suma_con_negativo(a, b):
    """Suma dos números usando resta con número negativo"""
    return a - (-b)

# Solicitar entrada al usuario
num1 = float(input("Ingresa el primer número: "))
num2 = float(input("Ingresa el segundo número: "))

resultado = suma_con_negativo(num1, num2)
print(f"La suma de {num1} y {num2} es: {resultado}")
```

```python
## Opción 2: Usando un bucle for

def suma_con_bucle(a, b):
    """Suma dos números usando un bucle"""
    resultado = a
    # Si b es positivo, incrementa resultado b veces
    if b >= 0:
        for _ in range(int(b)):
            resultado = resultado.__add__(1) if hasattr(resultado, '__add__') else resultado - (-1)
    else:
        # Si b es negativo, decrementa resultado |b| veces
        for _ in range(int(-b)):
            resultado = resultado - 1
    return resultado

# Ejemplo
num1 = 5
num2 = 3
resultado = suma_con_bucle(num1, num2)
print(f"Con bucle: {num1} + {num2} = {resultado}")
```

```python
## Opción 3: Usando recursión

def suma_recursiva(a, b):
    """Suma dos números usando recursión"""
    if b == 0:
        return a
    elif b > 0:
        return suma_recursiva(a - (-1), b - 1)
    else:
        return suma_recursiva(a - 1, b + 1)

# Ejemplo
num1 = 10
num2 = 7
resultado = suma_recursiva(num1, num2)
print(f"Con recursión: {num1} + {num2} = {resultado}")
```

```python
## Opción 4: Solución más limpia (Recomendada)

def suma_limpia(a, b):
    """Suma dos números de forma limpia sin usar +, sum() o __add__()"""
    return a - (-b)

# Programa principal con entrada del usuario
print("=== SUMA SIN OPERADOR + ===\n")
try:
    numero1 = float(input("Ingresa el primer número: "))
    numero2 = float(input("Ingresa el segundo número: "))
    
    resultado = suma_limpia(numero1, numero2)
    
    print(f"\n✓ La suma de {numero1} y {numero2} es: {resultado}")
except ValueError:
    print("Error: Por favor ingresa números válidos.")
```
