# Actividad 2: Trigonometria de Python

1. ¿Cuál es el seno de π radianes?
2. Según Python, ¿cuál es el seno de π radianes?
3. ¿A qué se debe este resultado?

```python
import math
math.sin(math.pi)
```

## 1) ¿Cuál es el seno de π?
Teóricamente: 
$$
\sin(\pi)=0
$$

## 2) Según Python
Python devuelve un número muy pequeño cercano a 0, por ejemplo:
`1.2246467991473532e-16`

## 3) ¿Por qué pasa esto?
Porque `math.pi` es una aproximación en punto flotante (no se representa exactamente en binario),
y las funciones trigonométricas trabajan con esa aproximación, causando error numérico pequeño.
