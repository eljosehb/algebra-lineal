# Cuadros M√°gicos
### Definimos dos cuadrados m√°gicos 3√ó3 ùëÄ y ùëÅ
### Aplicamos suma y producto por escalar y verificamos que siguen siendo m√°gicos


```python
import numpy as np

M = np.array([[8,1,6],
              [3,5,7],
              [4,9,2]])

N = np.ones((3,3), dtype=int)

S = M + N
T = 2 * M

def es_magico(A):
    s = A[0].sum()
    filas = all(A[i].sum() == s for i in range(3))
    cols  = all(A[:,j].sum() == s for j in range(3))
    diag1 = np.trace(A) == s
    diag2 = np.trace(np.fliplr(A)) == s
    return filas and cols and diag1 and diag2, s

print("M:", es_magico(M))
print("N:", es_magico(N))
print("M+N:", es_magico(S))
print("2M:", es_magico(T))
```
