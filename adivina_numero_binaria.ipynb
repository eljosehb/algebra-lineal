# Adivinador de Números con Búsqueda Binaria

Este programa:
1. Pide dos números distintos al usuario
2. Solicita que piense en un número entre esos dos
3. Adivina el número usando un algoritmo de búsqueda binaria
4. Solo contiene **una función**

```python
def adivinar_numero():
    """Función única que implementa la búsqueda binaria para adivinar un número"""
    print("=== ADIVINADOR DE NÚMEROS CON BÚSQUEDA BINARIA ===")
    print()
    
    # Paso 1: Obtener dos números distintos
    while True:
        try:
            a = int(input("Ingresa el primer número: "))
            b = int(input("Ingresa el segundo número: "))
            
            if a == b:
                print("❌ Los números deben ser distintos. Intenta de nuevo.\n")
                continue
            break
        except ValueError:
            print("❌ Por favor ingresa números válidos.\n")
            continue
    
    # Asegurar que a < b
    if a > b:
        a, b = b, a
    
    print(f"\nPiensa en un número entre {a} y {b} (inclusive).")
    print("Responde con 'mayor', 'menor' o 'correcto' según corresponda.\n")
    
    # Búsqueda binaria
    izquierda = a
    derecha = b
    intentos = 0
    
    while izquierda <= derecha:
        intentos += 1
        medio = (izquierda + derecha) // 2
        
        print(f"Intento {intentos}: ¿Es el número {medio}?")
        respuesta = input("Respuesta (mayor/menor/correcto): ").strip().lower()
        print(respuesta)
        
        if respuesta == "correcto":
            print(f"\n✓ ¡Adiviné tu número en {intentos} intentos! El número era {medio}")
            break
        elif respuesta == "mayor":
            print(f"   El número es mayor que {medio}\n")
            izquierda = medio + 1
        elif respuesta == "menor":
            print(f"   El número es menor que {medio}\n")
            derecha = medio - 1
        else:
            print("   Respuesta no válida. Intenta de nuevo con 'mayor', 'menor' o 'correcto'.\n")
            intentos -= 1
    
    if izquierda > derecha:
        print("\n⚠️ Parece que hay un problema con las respuestas proporcionadas.")

# Ejecutar el programa
adivinar_numero()
```
